```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img2 = cv2.imread('.\lenna.jpg')
color = ('b','g','r')
# enumerate():python里的一个新函数
# 它的作用：同时遍历索引(i)和元素(col)
for i,col in enumerate(color):
    histr = cv2.calcHist([img2], [i], None, [256], [0, 256])
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
plt.show()
```
